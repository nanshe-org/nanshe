```python
def generate_type_keys(hdf5_data, internal_path, type_op):
    for each_key in hdf5_data[internal_path]:
        try:
            each_value = type_op(each_key)
            yield( (each_key, each_value) )
        except ValueError:
            pass
```

```python
def generate_item_selection_list(hdf5_data_name, internal_path, selection_key):
    list_item_selection = []
    
    with h5.File(hdf5_data_name, 'r') as hdf5_data:
        for each_key in hdf5_data[internal_path]:
            if selection_key in hdf5_data[internal_path][each_key]:
                list_item_selection.append(hdf5_data[internal_path][each_key][selection_key][:])
            else:
                list_item_selection.append(None)
    
    return(list_item_selection)
```

```python
def generate_image_selection_array(hdf5_data_name, internal_path, selection_key, expected_frame_shape):
    list_image_selection = []
    
    with h5.File(hdf5_data_name, 'r') as hdf5_data:
        for each_key in hdf5_data[internal_path]:
            if selection_key in hdf5_data[internal_path][each_key]:
                assert(hdf5_data[internal_path][each_key][selection_key].shape == expected_frame_shape)
                
                list_image_selection.append(hdf5_data[internal_path][each_key][selection_key][:])
            else:
                list_image_selection.append(numpy.zeros(expected_frame_shape))
    
    image_selection_array = numpy.array(list_image_selection)
    
    return(image_selection_array)
```

```python
def array_shape(new_array):
    if new_array is not None:
        return(new_array.shape)
    else:
        return(tuple())
```

```python
neuron_masks_transformed = numpy.zeros(neurons["mask"].shape, dtype = float)

for i in xrange(len(neurons["mask"])):
    neuron_masks_transformed[i] = scipy.ndimage.morphology.distance_transform_edt(neurons["mask"][i] > 0)
```

```python
threshhold = 1

thickness = 1
half_thickness = thickness / 2

lower_threshhold = threshhold - half_thickness

upper_threshhold = threshhold + half_thickness


neuron_masks_transformed_threshhold = ( (lower_threshhold <= neuron_masks_transformed) & (neuron_masks_transformed <= upper_threshhold) )

neuron_masks_transformed_threshhold_max = (anp.expand_view(np.arange(1, 1 + len(neuron_masks_transformed_threshhold)), reps_after = neuron_masks_transformed_threshhold.shape[1:]) * neuron_masks_transformed_threshhold).sum(axis = 0)
```

```python
neuron_masks_transformed_threshhold_max_and_images_max = neuron_masks_transformed_threshhold_max.astype(float)
neuron_masks_transformed_threshhold_max_and_images_max /= (neuron_masks_transformed_threshhold_max_and_images_max.max() - neuron_masks_transformed_threshhold_max_and_images_max.min())
neuron_masks_transformed_threshhold_max_and_images_max *= (images_max.max() - images_max.min())
neuron_masks_transformed_threshhold_max_and_images_max += images_max
```

```python
neuron_masks_transformed_threshhold_max_and_dictionary_max = neuron_masks_transformed_threshhold_max.astype(float)
neuron_masks_transformed_threshhold_max_and_dictionary_max /= (neuron_masks_transformed_threshhold_max_and_dictionary_max.max() - neuron_masks_transformed_threshhold_max_and_dictionary_max.min())
neuron_masks_transformed_threshhold_max_and_dictionary_max *= (dictionary_max.max() - dictionary_max.min())
neuron_masks_transformed_threshhold_max_and_dictionary_max += dictionary_max
```
